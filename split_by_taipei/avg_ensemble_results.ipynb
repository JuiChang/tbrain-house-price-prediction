```python
import numpy as np
import pandas as pd
```

```python
xgb_ensmble_result = pd.read_csv('../output/submission_new_xgb_ensemble_split_by_taipei.csv')
# xgb_ensmble_result = pd.read_csv('../output/submission_new_xgb_area_valid_both.csv')
lgbm_ensmble_result = pd.read_csv('../output/submission_new_lgbm_ensemble_split_by_taipei.csv')
```

```python
sub_avg_ensmble_result = pd.concat([xgb_ensmble_result.set_index('building_id'), lgbm_ensmble_result.set_index('building_id')], axis=1, join='inner')
sub_avg_ensmble_result['avg'] = (sub_avg_ensmble_result.iloc[:, 0] + sub_avg_ensmble_result.iloc[:, 1]) / 2
```

```python
sub_avg_ensmble_result.shape
```

```python
submission = pd.DataFrame({
    "building_id": sub_avg_ensmble_result.index, 
    "total_price": sub_avg_ensmble_result.avg
})
submission.to_csv('../output/submission_new_correct_avg_ensemble_results_split_by_taipei.csv', index=False)
```
