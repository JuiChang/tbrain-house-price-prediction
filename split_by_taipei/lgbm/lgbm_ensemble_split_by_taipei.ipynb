```python
import numpy as np
import pandas as pd
```

```python
sub_in_tp = pd.read_csv('../../output/submission_lgbm_area_trainByAlltrain_in_taipei.csv')
sub_out_tp = pd.read_csv('../../output/submission_lgbm_area_trainByAlltrain_out_from_taipei.csv')
sub_all = pd.read_csv('../../output/submission_lgbm_area_trainByAlltrain_both.csv')
```

```python
sub_avg_in_tp = pd.concat([sub_all.set_index('building_id'), sub_in_tp.set_index('building_id')], axis=1, join='inner')

sub_avg_in_tp['avg'] = (sub_avg_in_tp.iloc[:, 0] + sub_avg_in_tp.iloc[:, 1]) / 2

# sub_avg_in_tp.head(10)
```

```python
sub_avg_out_tp = pd.concat([sub_all.set_index('building_id'), sub_out_tp.set_index('building_id')], axis=1, join='inner')

sub_avg_out_tp['avg'] = (sub_avg_out_tp.iloc[:, 0] + sub_avg_out_tp.iloc[:, 1]) / 2

# sub_avg_out_tp.head(10)
```

```python
sub_avg_split_by_tp = pd.concat([sub_avg_out_tp, sub_avg_in_tp])
```

```python
submission = pd.DataFrame({
    "building_id": sub_avg_split_by_tp.index, 
    "total_price": sub_avg_split_by_tp.avg
})
submission.to_csv('../../output/submission_lgbm_ensemble_split_by_taipei.csv', index=False)
```
